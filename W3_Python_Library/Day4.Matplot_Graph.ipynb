```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## III. Matplotlib Case Study - Graph

### III-1. 꺾은선 그래프 (Plot)

- `plt.plot()`
- 시계열 데이터 (x축 시간, y축 변수) - 이 전 시간에 비해 값이 얼마나 바뀌었는가

```python
x = np.arange(20)
y = np.random.randint(0, 20, 20) ## 난수를 20번 생성

# y축을 20 까지 보이게 하고 싶다면?
plt.axis([0, 20, 0, 20])

# y축을 "5"단위로 보이게 하고 싶다면?
plt.yticks([i for i in range(0, 21, 5)])

plt.plot(x, y)
plt.show()
```

### III-2. 산점도 (Scatter Plot)

- `plt.scatter()`
- x와 y가 별개의 변수일 때 사용. 별개인 변수에서 (상관)관계를 볼 수 있음

```python
plt.scatter(x, y)
plt.show()
```

### III-3. 박스 그림 (Box Plot)

- `plt.boxplot()`
- 단일 변수(수치형 데이터)의 분포를 볼 수 있음
- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)


- 두 개 이상의 변수를 함께 그려서 비교 -> 단일 인자 뿐만 아니라 Container를 이용해서 여러가지 인자 전달 가능

```python
plt.boxplot(y)

## Plot의 Title을 "Box Plot of y"라고 하고 싶다면?
plt.title('Box Plot of y')

plt.show()
```

```python
plt.boxplot((x, y))
plt.title('Box Plot of x, y')
plt.show()
```

### III-4. 막대 그래프 (Bar Plot)

- `plt.bar()`
- 범주형 데이터의 "값"과 그 크기를 직사각형으로 나타낸 그림
- 두 막대 사이에 간격이 있음

```python
plt.bar(x, y)

# xtics를 올바르게 처리해 봅시다.
plt.xticks(np.arange(0, 20, 1))

plt.show()
```

### III-5. 히스토그램 (Histogram)
- `plt.hist(data,bins=)`
- **도수** 분포표를 직사각형의 막대 형태로 나타냈다. (막대 그래프와 비슷)
- **"계급"**으로 나타낸 것이 특징 : 0, 1, 2가 아니라 0-2까지의 "범주형" 데이터로 구성 후 그림을 그림
- `bins`가 범주를 나타냄
- 두 막대 사이에 간격이 없음. 값이 연속적이라 가정하기 때문

```python
plt.hist(y, bins=np.arange(0, 20, 2))

# xtics를 올바르게 고쳐봅시다.
plt.xticks(np.arange(0, 20, 2))

plt.show()
```

### III-6. 원형 그래프 (Pie Chart)

- `plt.pie()`
- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 **비율** 확인에 용이

- 데이터에 대응되는 label을 붙여주는 것이 좋음!

```python
z = [100, 300, 200, 400]

plt.pie(z, labels=['one', 'two', 'three', 'four'])
plt.show()
```
