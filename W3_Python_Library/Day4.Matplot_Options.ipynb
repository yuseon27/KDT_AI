## II. Matplotlib Case Study

### II - 1. Case Study with Arguments

#### 1. 그래프 Plot 하기
* Plot : `plt.plot(graph)`
* plt 확인, 보여주기 `plt.show()`

```python
plt.plot([1, 2, 3, 4, 5])   ## 실제 plotting하는 함수 : Default 꺾은선 그래프  # y = x + 1 
plt.show()                  # plt를 확인하는 명령
```

```python
plt.plot([2, 4, 2, 4, 3])
plt.show()
```

#### 2. Figsize : Figure(도면)의 크기 지정하기
* `plt.figure(figsize=(a, b))`

```python
plt.figure(figsize=(6, 6))   ## plotting을 할 도면을 선언
plt.plot([0, 1, 2, 3, 4])    ## 실제 plotting을 하는 함수 # y = x
plt.show()                   ## plt를 확인하는 명령
```

```python
plt.figure(figsize=(3, 3))   ## plotting을 할 도면을 선언
plt.plot([0, 1, 2, 3, 4])    ## 실제 plotting을 하는 함수 # y = x
plt.show()                   ## plt를 확인하는 명령
```

### II - 2. 2차 함수 그래프 with plot()

```python
# 리스트를 이용해서 1차함수 y=x를 그려보면 :
plt.plot([0, 1, 2, 3, 4])    ## 실제 plotting을 하는 함수 # y = x
plt.show()                   ## plt를 확인하는 명령
```

```python
# numpy.array를 이용해서 함수 그래프 그리기
# y=x^2

x = np.array([1, 2, 3, 4, 5]) # 정의역
y = np.array([1, 4, 9, 16, 25]) # f(x)

plt.plot(x, y)  ## Sparse, 꺽은선 그래프
plt.show()
```

```python
# np.arange(start, end, c  -10 ~ 10 : c // c : 0.01

x = np.arange(-10, 10, 0.01)
plt.plot(x, x**2)
plt.show()
```

#### 3. X축, Y축에 설명 추가하기
* X축 : `plt.xlabel('x-axis')`
* Y축 : `plt.ylabel('y-axis')`

```python
x = np.arange(-10, 10, 0.01)

###
plt.xlabel('x value')
plt.ylabel('f(x) value')
###

plt.plot(x, x**2)
plt.show()
```

#### 4. X축, Y축의 범위를 설정하기
`plt.axis([x_min, x_max, y_min, y_max])`

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel('x value')
plt.ylabel('f(x) value')

###
plt.axis([-5, 5, 0, 25])  # [x_min, x_max, y_min, y_max]
###

plt.plot(x, x**2)
plt.show()
```

#### 5. X축, Y축에 눈금 설정하기 
- X축 : `plt.xticks(x_values' container)`
- Y축 : `plt.yticks(y_values' container)`

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25])  # [x_min, x_max, y_min, y_max]

###
plt.xticks([i for i in range(-5, 5, 1)])  # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0, 27, 3)])  # y축의 눈금 설정 
###

plt.plot(x, x**2)
plt.show()
```

#### 6. 그래프에 Title 달기
`plt.title("title_content")`

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25])  # [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5, 5, 1)])  # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0, 27, 3)])  # y축의 눈금 설정 

###
plt.title("y = x^2 Graph")
###

plt.plot(x, x**2)
plt.show()
```

#### 7. 범례 그리기
`plt.legend()`
- `legend()`는 plt.plot()에서 `label`을 지정해 준 후 사용해야함!

```python
x = np.arange(-10, 10, 0.01)

plt.xlabel('x value')
plt.ylabel('f(x) value')

plt.axis([-5, 5, 0, 25])  # [x_min, x_max, y_min, y_max]

plt.xticks([i for i in range(-5, 5, 1)])  # x축의 눈금 설정, -5, -4, -3, ...
plt.yticks([i for i in range(0, 27, 3)])  # y축의 눈금 설정 

plt.title("y = x^2 Graph")

###
plt.plot(x, x**2, label="trend")
plt.legend()
###

plt.show()
```
